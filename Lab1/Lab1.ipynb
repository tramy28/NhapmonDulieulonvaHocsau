```python
import pandas as pd
import numpy as np
```

```python
#Đọc dữ liệu đầu vào
data = pd.read_csv("dulieuxettuyendaihoc.csv", index_col= "STT")
    #read_csv() : hàm trong pandas để đọc dữ liệu từ file csv và trả về một DataFrame
    #index_col : một tham số của hàm, cho biết cột nào trong file được dử dụng làm chỉ số index của DataFrame
```

```python
# Câu 4: Thống kê dữ liệu thiếu trong cột DT trước khi thay thế:
data["DT"].isnull().sum()
    # data["DT"]: truy cập vào cột cụ thể
    # insnull(): xác định mỗi phần tử trong cột có phải giá trị thiếu không Kết quả trả về một 
    #   Series, trong đó mỗi phần tử là True nếu tương ứng với phần tử trong cột là giá trị thiếu
    #   (NaN), và False nếu không phải.'''
    # sum() :  tính tổng của các giá trị trong Series. Trong trường hợp này, True được coi là 1 và False được coi là 0
```

```python
# Câu 4: Thay thế các giá trị thiếu bằng 0:
data["DT"].fillna(0, inplace = False)
#       inplace = True -- Muốn thay đổi dữ liệu gốc
#       inplace = False -- Không muốn thay đổi dữ liệu gốc 
```

```python
# Câu 4: Số lượng giá trị thiếu sau khi thay thế: 
data["DT"].isnull().sum()
```

```python
# Câu 5: Thống kê dữ liệu thiếu cho biến T1
data["T1"].isnull().sum()
```

```python
# Câu 5: Hiệu chỉnh dữ liệu
data["T1"].fillna(data["T1"].mean(), inplace= False)
    # mean(): được gọi trên cột "T1", trả về giá trị trung bình của cột đó.
    # fillna(): sử dụng để điền các giá trị thiếu bằng một giá trị đã được chỉ định
```

```python
# Câu 6: Xử lý lần lượt tất cả các dữ liệu thiếu cho các biến về điểm số
is_float64 = data.dtypes == float
    # data.dtypes : trả về một Series chứa kiểu dữ liệu của mỗi cột
    # == float: so sánh kiểu dữ liệu của mỗi cột với kiểu dữ liệu float
    # is_float64: là một Series boolean với các giá trị True và False
data_column_list = data.columns[is_float64].tolist()
    # data.columns: Trả về một Index (mảng) chứa các tên cột trong DataFrame data
    # [is_float64]: Dùng để lựa chọn các tên cột tương ứng với các vị trí có giá trị True trong Series 
    # .tolist(): Chuyển đổi Index thành danh sách Python
```

```python
data_column_list
```

```python
for col in data_column_list:
    data[col].fillna(data[col].mean(), inplace=False)
```

```python
data.head()
```

```python
# Câu 7: Tạo các biế TBM1, TBM2, TBM3 và tính điểm theo công thức tính
# TBM = (T*2 + L + H + S + V*2 + X + D + N) / 10
for i in ["1", "2", "3"]:
    data["TBM"+i]=(data["T"+i]*2+ data["L"+i]+ data["H"+i]+data["S"+i]+data["V"+i]*2+data["X"+i]+data["D"+i]+data["N"+i])/10
```

```python
# Câu 7:
data.head()
```

```python
# Câu 8: Tạo hàm áp dụng điều kiện
def XL(DTBM):
    if DTBM < 5.0:
         return 'Y'  
    elif 5.0 <= DTBM < 6.5:
        return 'TB'
    elif 6.5 <= DTBM < 8.0:
        return 'K'  
    elif 8.0 <= DTBM < 9.0:
        return 'G'  
    else:
        return 'XS' 
```

```python
# Câu 8: Tạo vòng lặp để tạo biến xếp loại
for i in ["1", "2", "3"]:
    data["XL"+i]=(data["TBM"+i].apply(lambda DTBM :XL(DTBM)))
```

```python
# Câu 8: 
data.head()
```

```python
# Câu 9: Tìm giá trị tối thiểu và tối đa của điểm trung bình lớp 10, 11, 12
Diem_min = data[['TBM1', 'TBM2', 'TBM3']].min().min()
    # min(): Phương thức này trả về giá trị tối thiểu cho mỗi cột trong DataFrame con đã trích xuất
    #   Có nghĩa là nó sẽ trả về một Series chứa giá trị tối thiểu của "TBM1", "TBM2" và "TBM3"
    # min(): Sau khi đã có Series chứa giá trị tối thiểu của các cột, chúng ta gọi lại phương thức .min()
    #   để trả về giá trị tối thiểu của tất cả các giá trị tối thiểu đó. Đảm bảo rằng chúng ta có giá trị 
    #   tối thiểu nhất trong tất cả các điểm trung bình từ các năm lớp 10, 11 và 12
Diem_max = data[['TBM1', 'TBM2', 'TBM3']].max().max()
```

```python
# Câu 9: Tạo vòng lặp áp dụng Min-Max Normalization để chuyển điểm trung bình từ thang điểm 10 sang thang điểm 4
for i in ["1", "2", "3"]:
    data["US_TBM"+i]=((data["TBM"+i]-Diem_min)/(Diem_max -Diem_min))*4
```

```python
# Câu 9:
data.head()
```

```python
# Câu 10: Tạo biến KQXT mặc định, nếu kết quả= 0: rớt, nếu kq = 1 :đậu
data['KQXT'] = None
```

```python
# Câu 10: Xác định điều kiện cho từng khối và cập nhật biến KQXT tương ứng
# Duyệt qua từng sinh viên và xác định kết quả xét tuyển và khối
for index, row in data.iterrows():
    # Tính điểm trung bình theo khối
    diem_A_A1 = (row['DH1']*2 + row['DH2'] + row['DH3']) / 4
    diem_B = (row['DH1'] + row['DH2']*2 + row['DH3']) / 4
    diem_khac = (row['DH1'] + row['DH2'] + row['DH3']) / 3

    # Kiểm tra nếu một trong ba điểm trung bình lớn hơn hoặc bằng 5.0
    if diem_A_A1 >= 5.0:
        data.at[index, 'KQXT'] = 1
    elif diem_B >= 5.0:
        data.at[index, 'KQXT'] = 1  
    elif diem_khac >= 5.0:
        data.at[index, 'KQXT'] = 1
```
